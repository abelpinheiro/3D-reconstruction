```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# Carregar os dados da calibração da câmera
ret = np.load('./camera_params/ret.npy')
K = np.load('./camera_params/K.npy')
dist = np.load('./camera_params/dist.npy')

# Path das imagens
img_path1 = './reconstruct_this/alcool-left.png'
img_path2 = './reconstruct_this/alcool-right.png'

# Carregar imagens 
img_1 = cv2.imread(img_path1)
img_2 = cv2.imread(img_path2)

img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)

# Pegar altura e largura. Precisa ser o mesmo tamanho
#h,w = img_2.shape[:2]
w, h = np.shape(img_1)
map_d = np.zeros((w,h), np.uint8)
```

```python
list_ssd = []
for x_linha in range(aux, coords_x - aux):
    temp = 0
    ssd = 0
    for j in range(-aux, aux):
        for i in range(-aux, aux):
            temp = int(img_1[y+j][x+i]) - int(img_2[y+j][x_linha+i])
            ssd += temp * temp
    list_ssd.append(ssd)
```

```python
def get_disparity_map(img_1, img_2, kernel, map_d, ratio):
    aux = kernel//2
    print('aux: ', aux)
    coords_y, coords_x = map_d.shape
    for y in range(aux, coords_y - aux):
        if y % 10 == 0:
            print("y: ", y)
        for x in range(aux, coords_x - aux):
            list_ssd = np.repeat(np.inf, coords_x)
            min_ssd = np.inf
            disparity = 0
            
            if(ratio>= 0):
                ini = x - ratio
                if(ini<=aux):
                    ini = aux
                rango = range(ini,x)
            elif(ratio<0):
                fin = x - (ratio)
                if(fin>=coords_x - aux):
                    fin = coords_x - aux
                rango = range(x,fin)
            for x2 in rango:
                ssd = 0
                temp = 0
                ssd = np.subtract(img_1[y-aux:y+aux+1, x-aux:x+aux+1], \
                                 img_2[y-aux:y+aux+1, x2-aux:x2+aux+1])
                list_ssd[x2] = np.sum(np.multiply(ssd,ssd))
            map_d[y,x] = abs(x - np.argmin(list_ssd))
```

```python
def filtro_da_mediana(kernel, img):
    aux = kernel // 2
    coords_y, coords_x = img.shape
    result = np.copy(img)
    for y in range(aux, coords_y - aux):
        for x in range(aux, coords_x - aux):
            result[y,x] = np.median(img[y-aux:y+aux+1, x-aux:x+aux+1])
    return result
```

```python
kernel = 3
best_dist = -25
get_disparity_map(img_1, img_2, kernel, map_d, best_dist)
```

```python
# Função para reduzir uma imagem por um fator
def downsample_image(image, reduce_factor):
    for i in range(0,reduce_factor):
        # Checagem se imagem é colorida ou em escala de cinza
        if len(image.shape) > 2:
            row,col = image.shape[:2]
        else:
            row,col = image.shape

        image = cv2.pyrDown(image, dstsize= (col//2, row // 2))
    return image



# Obter matriz de camera otimizada para melhor distorção 
new_camera_matrix, roi = cv2.getOptimalNewCameraMatrix(K,dist,(w,h),1,(w,h))

# distorcer as imagens
img_1_undistorted = cv2.undistort(img_1, K, dist, None, new_camera_matrix)
img_2_undistorted = cv2.undistort(img_2, K, dist, None, new_camera_matrix)

# Reduzir o tamanho da imagem para melhorar a tunagem do algoritmo
img_1_downsampled = downsample_image(img_1_undistorted,3)
img_2_downsampled = downsample_image(img_2_undistorted,3)
```

```python
def create_output(vertices, colors, filename):
	colors = colors.reshape(-1,3)
	vertices = np.hstack([vertices.reshape(-1,3),colors])

	ply_header = '''ply
		format ascii 1.0
		element vertex %(vert_num)d
		property float x
		property float y
		property float z
		property uchar red
		property uchar green
		property uchar blue
		end_header
		'''
	with open(filename, 'w') as f:
		f.write(ply_header %dict(vert_num=len(vertices)))
		np.savetxt(f,vertices,'%f %f %f %d %d %d')
```
